```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Lets Start with Basic Plot
xArgument = np.linspace(10,100,25)
yArgument = xArgument ** 2
plt.plot(xArgument,yArgument)
plt.xlabel("X -axis")
plt.ylabel("Y -axis")
plt.title("Title") 
plt.show()
```

```python
#Creating Multiple Figures in 1 Plot and Adding Legend
plt.plot(xArgument,yArgument,label='Economy')
plt.plot(xArgument,yArgument*2,label='GDP')
plt.legend(loc=6) #center left code 
plt.show()
```

```python
#creat multiple plots
plt.subplot(1,2,1) #Number of Rows, Column, Plot Number
plt.plot(xArgument,yArgument)

plt.subplot(1,2,2)
plt.plot(yArgument, xArgument,'r')
```

```python
#Object Orieneted Method

fig = plt.figure() #Canvas
axes=fig.add_axes([0,0,1,1])
axes.plot(xArgument,yArgument)
axes.set_title("Larger Plot")

#Percentage from 0-1 to deviate from Bottom Left 20%, Bottom upper 50%, 40% of canvas size width, 30% of canvas size length
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(yArgument,xArgument)
axes2.set_title("Smaller Plot")
```

```python
#Another Method
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3) #2 row, 3 subplots on each row
plt.tight_layout() #Fixing overlap of plots
axes[0].plot(yArgument,xArgument)
axes[1].plot(xArgument,yArgument)
plt.show()
```

```python

fig,axes = plt.subplots(nrows=2,ncols=3) #2 row, 3 subplots on each row
plt.tight_layout() #Fixing overlap of plots
axes[0,1].plot(yArgument,xArgument)
axes[1,2].plot(xArgument,yArgument)
plt.show()
```

```python
fig.savefig("myPlot.png",dpi=200) #Saves your figure
```

```python
#Plot Appearence
fig,axes = plt.subplots(nrows=1,ncols=3)
axes[0].plot(yArgument,xArgument,color="purple",linewidth=3,linestyle="--")
axes[1].plot(yArgument,xArgument,color="purple",linewidth=3,alpha=0.5) #Alpga for transperency
axes[2].plot(yArgument,xArgument,color="#FF8C00",linestyle="dotted",marker='o',markersize=4,markerfacecolor="black")
# linewidth
#Alpga for transperency
#rgb Hex Code for Custom colors
#rgb Hex Code for Custom colors
```

## Seaborn

#### Distribution Plots

```python
import seaborn as sns
sns.set_style('whitegrid')
sns.set_context(context="notebook")
#Matplotlib color map for palette colors when choosing Hue
#https://matplotlib.org/stable/users/explain/colors/colormaps.html
```

```python
data= sns.load_dataset('tips')
data
```

```python
#Univariate Distribution plot
sns.distplot(data["total_bill"],kde=True,bins=30,color="green")
#A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, 
# analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.
```

```python
sns.rugplot(data["total_bill"]) #Draw a line for every single datapoint
```

```python
sns.jointplot(x="total_bill",y="tip",data=data,kind="scatter")
```

```python
#BiVariate Distribution plot
sns.jointplot(x="total_bill",y="tip",data=data,kind="hex") #By Default scatter
```

```python
sns.jointplot(x="total_bill",y="tip",data=data,kind="kde",shade=True)
```

```python
sns.pairplot(data,hue="sex",palette='coolwarm')#Do jointplot for every possible numerical column combination, Differentiaite on basic of categories of sex
```

##### Categorical Plots

```python
#Categorical Plots
#X is categorical #Y is numerical
sns.barplot(x="sex",y="total_bill",data=data) #Looking at total bill mean for different categories of sex
```

```python
sns.countplot(x='sex',data=data) #Same as barplot but only Y axis chosen already to be count for X-axis categories
```

```python
sns.boxplot(x='total_bill',data=data) #X for category
```

```python
sns.boxplot(x='day',y="total_bill",data=data,hue="smoker") #X for category
```

```python
sns.violinplot(x='day',y="total_bill",data=data,hue="smoker") #It gives more information 
```

```python
sns.violinplot(x='day',y="total_bill",data=data,hue="smoker",split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=data,palette='Set1',hue="sex") #Scatter plot based on category
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='violin')#General method where you define what kind of plot you need
```
